### Yolo model v8s

Object detection by YOLOv8
1. Load the YOLOv8 model
2. Load the images
3. Train the model



```python
# Load the YOLOv8 model
model = YOLO("yolov8s.pt")
```

```python
results = model.train(
    data=data_path,
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    augment=True,  # Enable YOLO's built-in augmentations
    degrees=15.0,  # Rotation (±15 degrees)
    scale=0.5,     # Scaling (up to 50%)
    flipud=0.5,    # Vertical flip (50% chance)
    fliplr=0.5,    # Horizontal flip (50% chance)
    mosaic=0.5,    # Mosaic augmentation (50% chance)
    hsv_h=0.015,   # Hue variation (instead of color_jitter)
    hsv_s=0.7,     # Saturation variation
    hsv_v=0.4,     # Value (brightness) variation
    cls=0.5,       # Classification loss weight (global, not per-class)
    patience=20,   # Early stopping to avoid overfitting
    project='/content/drive/MyDrive/yolo_training_results_01', # Save to Google Drive
    name='train_enhanced',
    exist_ok=True
)
```

# YOLOv10

```python
model10s = YOLO("yolov10s.pt")
```

```python
results = model10s.train(
    data=data_path,
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    augment=True,  # Enable YOLO's built-in augmentations
    degrees=15.0,  # Rotation (±15 degrees)
    scale=0.5,     # Scaling (up to 50%)
    flipud=0.5,    # Vertical flip (50% chance)
    fliplr=0.5,    # Horizontal flip (50% chance)
    mosaic=0.5,    # Mosaic augmentation (50% chance)
    hsv_h=0.015,   # Hue variation (instead of color_jitter)
    hsv_s=0.7,     # Saturation variation
    hsv_v=0.4,     # Value (brightness) variation
    cls=0.5,       # Classification loss weight (global, not per-class)
    patience=20,   # Early stopping to avoid overfitting
    project='runs/detect',
    name='trainv10s',
    exist_ok=True
)
```

```python
# Load the trained model
modelv10s = YOLO('/content/runs/detect/trainv10s/weights/best.pt')
```

```python
# Evaluate the model
results = modelv10s.val(data=data_path)
print(f"mAP@50: {results.box.map50:.3f}")
print(f"mAP@50:95: {results.box.map:.3f}")
```

```python
# Predict on new images
results = modelv10s.predict(
    source='/content/IMG_9370 (1).jpg',  # Test Image
    conf=0.37,                # confidence threshold
    save=True                # saves to runs/detect/predict
)
```

```python
# Analyze predictions for impurity vs. purity percentage
total_detections = 0
impurity_count = 0
purity_count = 0

for result in results:
    boxes = result.boxes.xyxy
    cls = result.boxes.cls
    for c in cls:
        total_detections += 1
        if c == 0:  # Impurity
            impurity_count += 1
        elif c == 1:  # Pure-Black-Pepper
            purity_count += 1

impurity_percentage = (impurity_count / total_detections) * 100 if total_detections > 0 else 0
purity_percentage = (purity_count / total_detections) * 100 if total_detections > 0 else 0

print(f"Impurity Percentage: {impurity_percentage:.2f}%")
print(f"Purity Percentage: {purity_percentage:.2f}%")
```
